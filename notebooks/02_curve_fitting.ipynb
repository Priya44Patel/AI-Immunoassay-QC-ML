```python
from src.preprocessing import process_raw_folder, aggregate_replicates
from src.curve_fit import fit_4pl, generate_curve
import matplotlib.pyplot as plt

df = process_raw_folder("data/raw")
agg = aggregate_replicates(df)

stds = agg[agg["Type"] == "STD"]

params, r2 = fit_4pl(stds["Concentration"], stds["Signal_mean"])
print("R²:", r2)

x, y = generate_curve(params)
plt.plot(x, y)
plt.scatter(stds["Concentration"], stds["Signal_mean"])
plt.xscale("log")
plt.show()
```
