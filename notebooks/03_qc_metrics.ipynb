```python
from src.preprocessing import process_raw_folder, aggregate_replicates, save_processed
from src.qc_utils import compute_cv, compute_recovery, flag_outliers

df = process_raw_folder("data/raw")
agg = aggregate_replicates(df)

agg["CV"] = agg["Signal_std"] / agg["Signal_mean"] * 100
agg["Recovery"] = agg.apply(
    lambda r: compute_recovery(r["Signal_mean"], r["Concentration"]), axis=1
)
agg["Outlier"] = flag_outliers(agg["Signal_mean"])

save_processed(agg)
agg.head()
```
